```python
x = np.arange(0,10)
```

```python
y = x**2
```

```python
y
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'*')
```

```python
plt.plot(x,y,'b*')
```

```python
plt.plot(x,y,'r--')
plt.xlim(2,4)
plt.ylim(3,20)
plt.title("Zoomed")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```
